```python
def Mu(df,tradDays):
    mu  = df.mean()*tradDays
    
    mu.values.T
    return mu

def Sigma(df,tradDays):
    sigma = df.cov()*tradDays
    
    sigma = sigma.values
    return sigma
    
```

```python
def Delta(df):
    delta = np.full((len(df.columns),1),0.02)
    delta[len(df.columns)-1] = 0.005
    return delta    
```

```python
import cvxpy as cp
```

```python
train =stockret.loc["2000-01-01":"2016-12-31"]
val = stockret.loc["2017-01-01":"2018-12-31"]
test = stockret.loc["2019-01-01":"2019-06-31"]
```

```python
def optimize(train,tradDays,x = 1,lambda0 = 1,give = "w"):
    w = cp.Variable(len(train.columns))
    w_plus = cp.Variable(len(train.columns))
    w_minus = cp.Variable(len(train.columns))
    delta = x*Delta(train).reshape(len(train.columns),1)

    ret = Mu(train,tradDays).to_numpy().T*w
    aux = delta.T*(w_plus + w_minus)
    risk = cp.quad_form(w,Sigma(train,tradDays))

    prob = cp.Problem(
        cp.Minimize(risk-lambda0*ret + lambda0*aux),
        [sum(w) ==1,
        w == w_plus - w_minus,
        w_plus>=0,
        w_minus>=0,
        w>=0])

    prob.solve()
    param = []
    param.append(w.value)
    param.append(Mu(train,tradDays).dot(w.value))
    param.append(np.sqrt(w.value.dot(Sigma(train,tradDays)).dot(w.value)))
    
    if give == "w":
        return w.value
    elif give == "return":
        return Mu(train,tradDays).dot(w.value)
    elif give == "std":
        param.append(Mu(train,tradDays).dot(w.value))
        return np.sqrt(w.value.dot(Sigma(train,tradDays)).dot(w.value))
    elif give == "list":
        return param

```

```python
optimize(train,251,give="list")[1]
```

```python
for i in np.arange(10,30,1):
    
    t = optimize(train,251,i,1,"list")
    print(str(i)+" "+str(t[1])+" "+str(t[2]))
```
