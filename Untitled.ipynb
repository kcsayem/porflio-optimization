```python
import pandas as pd 
import cvxpy as cp 
import numpy as np 
import sys 
import os.path
import warnings
warnings.filterwarnings('ignore') 

# backtesting portfolio

from empyrial import *
```

```python
tickers = pd.read_pickle('Data/DowJonesTicker.pickle')
modpath = os.path.dirname(os.path.abspath(sys.argv[0]))
temp = 'Data/Stocks/dow_jones/' + tickers[0] + '.csv'
datapath = os.path.join(modpath, temp)
data = pd.read_csv(temp, index_col='Date', parse_dates=True)
stock_df = pd.DataFrame()
stock_df[tickers[0]] = data['Close']

for i in range(1, len(tickers)):
    
    temp = 'Data/Stocks/dow_jones/' + tickers[i] + '.csv'
    datapath = os.path.join(modpath, temp)
    data = pd.read_csv(temp, index_col='Date', parse_dates=True)
    stock_df[tickers[i]] = data['Close']

```

```python
# Removing 'DOW' as it does not have data in all period
stock_df = stock_df.drop(['DOW'], axis=1)
```

```python
tnx_df = pd.read_csv('Data/Bonds/^TNX.csv',  index_col='Date', parse_dates=True)['Close']
bond = pd.DataFrame()
bond['bond'] = tnx_df
# asset_df = stock_df.copy()
stock_df = stock_df.loc['2009-01-01':]
asset_df=pd.concat([stock_df,bond], join='inner', axis=1)
asset_df.head()
```

```python
returns = np.log(asset_df/asset_df.shift(1))
returns.dropna(inplace=True)
returns
```

```python
train =  returns.loc['2009-01-01':'2017-12-31']
test =  returns.loc['2018-01-01':'2018-12-31']
final =  returns.loc['2019-01-01':'2019-12-31']
```

```python
def calc_Mu(df):
    mu = df.mean() * 252
    return mu

def calc_sigma(df):
    sigma = df.cov() * 252
    sigma = sigma.values
    return sigma
    
def calc_delta(df):
    delta = np.full((len(df.columns), 1), 0.02)
    delta[len(df.columns)-1] = 0.005
    return delta 
```

```python
def markwoitz_opt(df, lambda0 = 0.001):
    w = cp.Variable(len(df.columns))
    lambda0 = lambda0

    mu = calc_Mu(df).to_numpy()
    ret = mu.T * w
    Sigma = calc_sigma(df)
    risk = cp.quad_form(w, Sigma)

    prob = cp.Problem(
        cp.Minimize(risk - lambda0 * ret),
        [sum(w) == 1, w >= 0]
    )
    prob.solve()

    output = {'weights':w.value, 'total_weights':np.sum(w.value),
                'expected_return': mu.dot(w.value), 'std': np.sqrt(w.value.dot(Sigma).dot(w.value)),
                'stocks':np.sum(w.value[:-1]), 'bonds':w.value[-1]}
    return output
```

```python
res = markwoitz_opt(train, 0.02)
```

```python
for key, value in res.items():
    print(key, ":", value)
```

```python
def robust_opt(df, tuner = 1,  lambda0 = 0.02):
    w = cp.Variable(len(df.columns))
    w_plus = cp.Variable(len(df.columns))
    w_minus = cp.Variable(len(df.columns))
    delta = tuner * calc_delta(df).reshape(len(df.columns), 1)

    mu = calc_Mu(df).to_numpy()
    ret = mu.T * w
    aux = delta.T * (w_plus + w_minus) 
    Sigma = calc_sigma(df)
    risk = cp.quad_form(w, Sigma)

    prob = cp.Problem(
        cp.Minimize(risk - lambda0 * ret + lambda0 * aux),
        [   sum(w) == 1,
             w >= 0,
            w == w_plus - w_minus,
            w_plus >= 0,
            w_minus >= 0,
        ]
    )
    prob.solve()

    output = {'weights':w.value, 'total_weights':np.sum(w.value),
                'expected_return': mu.dot(w.value), 'std': np.sqrt(w.value.dot(Sigma).dot(w.value)),
                'stocks':np.sum(w.value[:-1]), 'bonds':w.value[-1]}
    return output
```

```python
res = robust_opt(train, 5)
res['expected_return']
```

```python
lambdas = np.arange(0.1, 3, 0.1)
tuners = np.arange(0.1, 3, 0.1)

optval = []
for tuner in tuners:
    for lamb in lambdas:
        res = robust_opt(train, tuner = tuner, lambda0 = lamb)
#         print('For lambda {} & tuner {} expected return {} & standard deviation {}'.format(lamb, tuner, res['expected_return'], res['std']))
        optval.append([lamb, tuner, res['expected_return'], res['std']])
        
rob_opt_df = pd.DataFrame(optval, columns=['lambda', 'tuner', 'expected return', 'standard deviation'])
```

```python
rob_opt_df.sort_values(['expected return', 'standard deviation'],
                          ascending=[True, False])
```
